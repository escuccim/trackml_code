```python
R=2000 #Helix Radius

w=np.pi/3 #length of arc

v=3000 # velocity in Z

theta0=-0.3 # angle at the origin

D=0 #D=500 - Use for helixes far from the origin

z0=0 #z0=200 - Z shifting

t=np.random.rand(100)

X=-R*np.sin(w*t+theta0)+(R-D)*np.sin(theta0)

Y=-R*np.cos(w*t+theta0)+(R-D)*np.cos(theta0)

Z=z0+v*t
```

```python
hits = pd.DataFrame()
hits['x'] = X
hits['y'] = Y
hits['z'] = Z
```

```python
dfh['stepped_z'] = dfh.z
dfh['r'] = np.sqrt(dfh['x'].values**2+dfh['y'].values**2+dfh['stepped_z'].values**2)
dfh['rt'] = np.sqrt(dfh['x'].values**2+dfh['y'].values**2)
dfh['a0'] = np.arctan2(dfh['y'].values,dfh['x'].values)
dfh['a2'] = np.arctan2(dfh['x'].values,dfh['y'].values)
dfh['z1'] = dfh['stepped_z'].values/dfh['rt'].values

dfh['z2'] = dfh['stepped_z']/dfh['r']
dfh['x2'] = dfh['x'].values/dfh['r'].values
dfh['y2'] = dfh['y'].values/dfh['r'].values

dfh['z3'] = np.arctan2(dfh['stepped_z'].values,dfh['rt'].values)
```

```python
cx = np.array([0.96, .96, .4, .05, 0.20, 0.0108, 0.0108, 1, 1])
# cx = np.array([1, 1, 1, 1, 1, 1, 1, 1, 1])
dfh['a1'] = dfh['a0']+1*(dfh['rt']+(0)*dfh['rt']**2)/1000*(1/3)/180*math.pi
dfh['sina1'] = np.sin(dfh['a1'].values)
dfh['cosa1'] = np.cos(dfh['a1'].values)
dfh['sina2'] = np.sin(dfh['a2'].values)
dfh['cosa2'] = np.cos(dfh['a2'].values)
ss = StandardScaler()
dfs = ss.fit_transform(dfh[['sina1','cosa1','z1','z2','z3','x2','y2', 'sina2', 'cosa2']].values)
dfs = np.multiply(dfs, cx)
```
